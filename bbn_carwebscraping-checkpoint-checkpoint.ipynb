```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import numpy as np
import re

# Start a Selenium WebDriver session
driver = webdriver.Chrome()

#city_code = [3686,2,4709,2378,5,1605]

city_code = [1605]
# Navigate to the webpage you want to scrape
for id in city_code:
    driver.get(f"https://www.cars24.com/buy-used-car?sort=bestmatch&serveWarrantyCount=true&listingSource=FilterTags&storeCityId={id}")
    # Set the total scrolls you want to perform
    total_scrolls = 5
    c = []

    # Auto-scrolling loop
    for i in range(total_scrolls):
        # Scroll down to the bottom of the page
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        # Wait for some time for the content to load (adjust as needed)
        time.sleep(60)

    # After scrolling, get the page source and pass it to BeautifulSoup
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')

    # Find elements using BeautifulSoup
    brand = []
    driven_km = []
    fuel_type = []
    types = []
    specs = []
    city = []
    price = []
    emi = []

    for i in soup.find_all("div", class_="_2YB7p"):
        text = i.find("h3", class_="_11dVb").text if i.find("h3", class_="_11dVb") else np.nan
        brand.append(text)

        text1 = i.find("ul",class_="_3J2G-").text if i.find("ul",class_="_3J2G-") else np.nan
        text1 = re.findall(r"\d+",text1)
        driven_km.append(text1)
        
        text2 = i.find("ul",class_="_3J2G-").text if i.find("ul",class_="_3J2G-") else np.nan
        text2 = re.findall(r"(Petrol|Diesel|CNG)",text2)
        fuel_type.append(text2)
        
        text3 = i.find("ul",class_="_3J2G-").text if i.find("ul",class_="_3J2G-") else np.nan
        text3 = re.findall(r"(Automatic|Manual)",text3)
        types.append(text3)

        text4 = i.find("span",class_="_3JoYA").text if i.find("span",class_="_3JoYA") else np.nan
        specs.append(text4)
        
        text5 = i.find("strong",class_="_3RL-I").text if i.find("strong",class_="_3RL-I") else np.nan
        text5 = re.findall(r"\d+.\d+",text5)
        price.append(text5)
         
        text6 = i.find("span",class_="_2O0yU").text if i.find("span",class_="_2O0yU") else np.nan
        text6 = re.findall(r"\d+,\d+",text6)
        emi.append(text6)
        
        
        

        if id == 3686:
            city.append("Hyderabad")
        elif id == 2:
            city.append("New Delhi")
        elif id == 5:
            city.append("Gurgaon")
        elif id == 4709:
            city.append("Bangalore")
        elif id == 2378:
            city.append("Mumbai")
        


    # Close the WebDriver session
    driver.quit()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({ "Brand":brand,
    "Driven_km":driven_km ,
    "Fuel_Type":fuel_type,
    "Type":types,
    "Feautures":specs,
    "City":city,
    "Price":price,
    "EMI_month":emi})
df
```
